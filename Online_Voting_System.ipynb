```python
import pandas as pd
import numpy as np
```

```python
voters_df = pd.read_excel("/content/Voterlist.xlsx")
candidates_df = pd.read_excel("/content/Candidate Datebase.xlsx")
if "Voted" not in voters_df.columns:
    voters_df["Voted"] = 0
if "Votes" not in candidates_df.columns:
    candidates_df["Votes"] = 0
print("Data Loaded Successfully ✅")
```

```python
def login():
    voter_name =input("Enter Name: ")
    password = input("Enter Password: ")
    voter = voters_df[voters_df["Name"] == voter_name]
    if voter.empty:
        print("Invalid name")
        return None
    if voter.iloc[0]["Password"] != password:
        print("Invalid Password ")
        return None
    print("Login Successful ")
    return voter_name
```

```python
def vote(voter_name):
    global voters_df, candidates_df
    voter_index = voters_df[voters_df["Name"] == voter_name].index[0]
    if voters_df.at[voter_index, "Voted"] == 1:
        print("You have already voted ")
        return
    print("\nCandidates List:")
    print(candidates_df[["Candidate ID", "Name", "Party"]])
    choice = input("Enter Candidate ID to vote: ")
    candidate = candidates_df[candidates_df["Candidate ID"] == choice]
    if candidate.empty:
        print("Invalid Candidate")
        return
    candidate_index = candidate.index[0]
    candidates_df.at[candidate_index, "Votes"] += 1
    voters_df.at[voter_index, "Voted"] = 1
    print("Vote Done Successfully")
```

```python
def show_results():
    print("\nFinal Results")
    print("-------------------------")
    print(candidates_df[["Name", "Party", "Votes"]])
```

```python
while True:
    print("\n1. Login & Vote")
    print("2. Show Results")
    print("3. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        voter = login()
        if voter:
            vote(voter)

    elif choice == "2":
        show_results()

    elif choice == "3":
        print("System Closed")
        break

    else:
        print("Invalid Choice")
```
